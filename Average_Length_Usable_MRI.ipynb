```python
from VSA_utilities import date_from_int, describe_lens, get_durs, time_rounding
from datetime import date
import numpy as np
import os
import pandas as pd
```

```python
unarchived_mri = "/Volumes/Data/Research/Healthy Brain Network/Voice Sample Data/Post-MRI Audio Video"
storages = {unarchived_mri, "/Volumes/Voice Sample Data"}
usable_table = pd.read_excel("AV participant list_112017.xlsx")
usable_table["Date of recording"] = usable_table["date_recorded"].apply(date_from_int)
```

```python
usable_table
```

```python
lens_usable = []
dur_dict = dict()
for storage in storages:
    usable, all_samples, dd, description = get_durs(storage, usable_table)
    lens_usable = [*lens_usable, *usable]
    for d in dd:
        dur_dict[d] = dd[d]
```

```python
for i in dur_dict:
    if(len(dur_dict[i]) > 1):
        print(max([dur_dict[i][j][0] for j in dur_dict[i]]))
```

```python
no_files_IDs = [x for x in usable_table['RandID'] if str(x) not in dur_dict]
no_files_IDs
```

```python
id_dur = dict()
randindex = usable_table['RandID'].astype(int)
for r in randindex:
    id_dur[r] = np.nan if str(r) not in dur_dict else list(dur_dict[str(r)].values())[0][1]
```

```python
len(id_dur)
```

```python
dur_frame = pd.DataFrame.from_dict(id_dur, orient='index').rename(columns={0: 'duration (seconds)'})
dur_frame.index.name = 'RandID'
```

```python
dur_table = usable_table.merge(dur_frame, left_on='RandID', right_index=True, how='inner')
```

```python
dur_table['RandID'] = dur_table['RandID'].astype(str)
```

```python
dur_table['RandID'][dur_table['RandID'].duplicated()]
```

```python
dur_table = dur_table[~dur_table['RandID'].duplicated()]
```

```python
dur_table = dur_table[~(dur_table['RandID'].astype(int).isin(no_files_IDs))]
```

```python
describe_lens(
    dur_table.as_matrix(
        columns=[
            'RandID',
            'Date of recording',
            'duration (seconds)'
        ]
    ), "usable"
)
```

```python
dur_table.to_csv(
    os.path.join(
        unarchived_mri,
        "Usable Voice Samples (with durations) {0}.csv".format(
            str(date.today())
        )
    ),
    index=False
)
```
