```python
from VSA_utilities import date_from_int, describe_lens, get_durs, time_rounding
from datetime import date
import numpy as np
import os
import pandas as pd
```

```python
unarchived_mri = "/Volumes/Data/Research/Healthy Brain Network/Voice Sample Data/Post-MRI Audio Video"
storages = {unarchived_mri, "/Volumes/Voice Sample Data"}
usable_table = pd.read_excel(os.path.join(unarchived_mri, "Usable Voice Samples.xlsx"))
usable_table["Date of recording"] = usable_table["Date of recording"].apply(date_from_int)
```

```python
lens_usable = []
dur_dict = dict()
for storage in storages:
    usable, all_samples, dd, description = get_durs(storage, usable_table)
    lens_usable = [*lens_usable, *usable]
    for d in dd:
        dur_dict[d] = dd[d]
```

```python
describe_lens(
    np.array([dur_dict[l][3] for l in dur_dict]),
    "all usable as of {0}".format(str(date.today()))
)
```

```python
print(np.sort(np.array([x for x in dur_dict])))
    #print(sum([dur_dict[x][r][1] for r in dur_dict[x]]))
```

```python
for x in usable_table['RandID']:
    if str(x) not in dur_dict:
        print(x)
```

```python
randindex = usable_table['RandID'].astype(int)
randindex = randindex.replace(5178472, 5174872)
```

```python
id_dur = dict()
for r in randindex:
    id_dur[r] = np.nan if str(r) not in dur_dict else list(dur_dict[str(r)].values())[0][1]
```

```python
dur_frame = pd.DataFrame.from_dict(id_dur, orient='index').rename(columns={0: 'duration (seconds)'})
dur_frame.index.name = 'RandID'
```

```python
dur_frame.index = pd.Series(dur_frame.index).replace(5174872, 5178472)
```

```python
dur_table = usable_table.merge(dur_frame, left_on='RandID', right_index=True, how='inner')
```

```python
dur_table['RandID'] = dur_table['RandID'].astype(str)
```

```python
dur_table['RandID'][dur_table['RandID'].duplicated()]
```

```python
dur_table = dur_table[~dur_table['RandID'].duplicated()]
```

```python
describe_lens(np.array(dur_table), "usable")
```

```python
dur_table.to_csv(
    os.path.join(unarchived_mri, "Usable Voice Samples (with durations).csv"),
    index=False
)
```
