```python
from VSA_utilities import date_from_int, describe_lens, get_durs, time_rounding
from datetime import date
import numpy as np
import os
import pandas as pd
```

```python
unarchived_mri = "/Volumes/Data/Research/Healthy Brain Network/Voice Sample Data/Post-MRI Audio Video"
storages = {unarchived_mri, "/Volumes/Voice Sample Data"}
usable_table = pd.read_excel(os.path.join(unarchived_mri, "Usable Voice Samples.xlsx"))
usable_table["Date of recording"] = usable_table["Date of recording"].apply(date_from_int)
```

```python
lens_usable = []
dur_dict = dict()
for storage in storages:
    usable, all_samples, dd, description = get_durs(storage, usable_table)
    lens_usable = [*lens_usable, *usable]
    for d in dd:
        dur_dict[os.path.join(dd[d][0], dd[d][1])] = dd[d]
```

```python
describe_lens(
    np.array([dur_dict[l][3] for l in dur_dict]),
    "all usable as of {0}".format(str(date.today()))
)
```

```python
time_rounding(np.array([dur_dict[l][3] for l in dur_dict]).sum())
```
